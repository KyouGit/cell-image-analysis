```python
from diffusers import StableDiffusionPipeline
import torch
import os
from PIL import Image
import pandas as pd
from tqdm import tqdm

# 생성할 데이터 설정
output_dir = "./generated_teardropcell"
os.makedirs(output_dir, exist_ok=True)

num_images = 100  # 생성할 이미지 수
prompt = "microscope image of a teardrop-shaped blood cell, low resolution, medical dataset"

# Stable Diffusion 파이프라인 로딩 (fp16은 GPU 권장)
pipe = StableDiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
    torch_dtype=torch.float16
).to("cuda")

# 생성된 이미지 메타데이터 저장용
records = []

# 이미지 생성 루프
for i in tqdm(range(num_images), desc="Generating images"):
    image = pipe(prompt).images[0]  # 생성된 이미지(PIL 객체)

    filename = f"teardrop_{i:04d}.png"
    filepath = os.path.join(output_dir, filename)
    image.save(filepath)

    # 메타데이터 기록
    records.append({
        "image": filepath,
        "label": "TearDropCell",
        "prompt": prompt
    })

# 메타데이터 CSV 저장
df = pd.DataFrame(records)
df.to_csv(os.path.join(output_dir, "metadata.csv"), index=False)
```
