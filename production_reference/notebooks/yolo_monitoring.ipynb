```python
import sys
sys.path.append('/home/smile/work/pbs06/yolov5')

def xywh2xyxy(x):
    # Convert nx4 boxes from [x, y, w, h] to [x1, y1, x2, y2]
    y = x.copy()
    y[..., 0] = x[..., 0] - x[..., 2] / 2  # x1
    y[..., 1] = x[..., 1] - x[..., 3] / 2  # y1
    y[..., 2] = x[..., 0] + x[..., 2] / 2  # x2
    y[..., 3] = x[..., 1] + x[..., 3] / 2  # y2
    return y

from utils.augmentations import letterbox, random_perspective, augment_hsv
from utils.dataloaders import Albumentations
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
import cv2
import yaml
import random

# hyp 로드
with open('/home/smile/work/pbs06/yolov5/data/hyps/hyp.scratch-low.yaml') as f:
    hyp = yaml.safe_load(f)

# 이미지와 라벨 경로
img_path = "/home/smile/work/pbs06/train/images/10001.jpg"
label_path = img_path.replace("/images/", "/labels/").replace(".jpg", ".txt")

img0 = cv2.imread(img_path)
h0, w0 = img0.shape[:2]
labels = np.loadtxt(label_path).reshape(-1, 5)  # class x y w h (normalized)

# 반복
num_augment = 6
fig, axes = plt.subplots(1, num_augment, figsize=(20, 4))

for i in range(num_augment):
    img = img0.copy()
    lbl = labels.copy()

    # Step 1: Resize + pad
    img, ratio, pad = letterbox(img, new_shape=384, auto=False, scaleup=True)
    lbl[:, 1:] = lbl[:, 1:] * [w0, h0, w0, h0]  # to absolute
    lbl[:, 1:] = xywh2xyxy(lbl[:, 1:])  # to xyxy
    lbl[:, 1:] = lbl[:, 1:] * [ratio[0], ratio[1], ratio[0], ratio[1]] + [pad[0], pad[1], pad[0], pad[1]]

    # Step 2: Perspective
    img, lbl = random_perspective(img, lbl,
                                  degrees=hyp['degrees'],
                                  translate=hyp['translate'],
                                  scale=hyp['scale'],
                                  shear=hyp['shear'],
                                  perspective=hyp['perspective'],
                                  border=(0, 0))

    # Step 3: Albumentations
    alb = Albumentations(size=384)
    img, lbl = alb(img, lbl)

    # Step 4: HSV
    augment_hsv(img, hgain=hyp['hsv_h'], sgain=hyp['hsv_s'], vgain=hyp['hsv_v'])

    # Step 5: Flip (xyxy 기준은 아래처럼)
    if random.random() < hyp['fliplr']:
        img = np.fliplr(img)
        img_w = img.shape[1]
        lbl[:, [1, 3]] = img_w - lbl[:, [3, 1]]

    if random.random() < hyp['flipud']:
        img = np.flipud(img)
        img_h = img.shape[0]
        lbl[:, [2, 4]] = img_h - lbl[:, [4, 2]]

    # 시각화
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    for box in lbl:
        cls, x1, y1, x2, y2 = box
        x1, y1, x2, y2 = map(int, [x1, y1, x2, y2])
        if x2 <= x1 or y2 <= y1:
            continue
        cv2.rectangle(img_rgb, (x1, y1), (x2, y2), (0, 255, 0), 2)

    axes[i].imshow(img_rgb)
    axes[i].set_title(f"Aug {i+1}")
    axes[i].axis("off")

plt.suptitle("라벨 박스 포함 증강 결과", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
import sys
sys.path.append('/home/smile/work/pbs06/yolov5')

from utils.augmentations import letterbox, random_perspective, augment_hsv
from utils.dataloaders import Albumentations
from utils.general import xywhn2xyxy
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
import cv2
import yaml
import random

def xywh2xyxy(x):
    y = x.copy()
    y[..., 0] = x[..., 0] - x[..., 2] / 2
    y[..., 1] = x[..., 1] - x[..., 3] / 2
    y[..., 2] = x[..., 0] + x[..., 2] / 2
    y[..., 3] = x[..., 1] + x[..., 3] / 2
    return y

# hyp 불러오기
with open('/home/smile/work/pbs06/yolov5/data/hyps/hyp.scratch-low.yaml') as f:
    hyp = yaml.safe_load(f)

img_path = "/home/smile/work/pbs06/train/images/10001.jpg"
label_path = img_path.replace("/images/", "/labels/").replace(".jpg", ".txt")

img0 = cv2.imread(img_path)
h0, w0 = img0.shape[:2]
labels = np.loadtxt(label_path).reshape(-1, 5)

# 증강 반복
num_augment = 6
fig, axes = plt.subplots(num_augment, 2, figsize=(8, 3 * num_augment))  # (행, 열)

for i in range(num_augment):
    img = img0.copy()
    lbl = labels.copy()

    # Step 1: Resize + pad
    img, ratio, pad = letterbox(img, new_shape=384, auto=False, scaleup=True)
    lbl[:, 1:] = lbl[:, 1:] * [w0, h0, w0, h0]
    lbl[:, 1:] = xywh2xyxy(lbl[:, 1:])
    lbl[:, 1:] = lbl[:, 1:] * [ratio[0], ratio[1], ratio[0], ratio[1]] + [pad[0], pad[1], pad[0], pad[1]]

    # Step 2: Perspective
    img, lbl = random_perspective(img, lbl,
                                  degrees=hyp['degrees'],
                                  translate=hyp['translate'],
                                  scale=hyp['scale'],
                                  shear=hyp['shear'],
                                  perspective=hyp['perspective'],
                                  border=(0, 0))

    # Step 3: Albumentations
    alb = Albumentations(size=384)
    img, lbl = alb(img, lbl)

    # 👇 복사본 만들기 (HSV 적용 전 이미지 저장)
    img_before = img.copy()

    # Step 4: HSV
    augment_hsv(img, hgain=hyp['hsv_h'], sgain=hyp['hsv_s'], vgain=hyp['hsv_v'])

    # Step 5: Flip
    if random.random() < hyp['fliplr']:
        img = np.fliplr(img)
        img_before = np.fliplr(img_before)
        img_w = img.shape[1]
        lbl[:, [1, 3]] = img_w - lbl[:, [3, 1]]

    if random.random() < hyp['flipud']:
        img = np.flipud(img)
        img_before = np.flipud(img_before)
        img_h = img.shape[0]
        lbl[:, [2, 4]] = img_h - lbl[:, [4, 2]]

    # 시각화용 RGB 변환
    img_before_rgb = cv2.cvtColor(img_before, cv2.COLOR_BGR2RGB)
    img_after_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # 박스 그리기
    for box in lbl:
        cls, x1, y1, x2, y2 = box
        x1, y1, x2, y2 = map(int, [x1, y1, x2, y2])
        if x2 <= x1 or y2 <= y1:
            continue
        cv2.rectangle(img_before_rgb, (x1, y1), (x2, y2), (255, 0, 0), 2)  # blue box (before)
        cv2.rectangle(img_after_rgb, (x1, y1), (x2, y2), (0, 255, 0), 2)   # green box (after)

    # 표시
    axes[i][0].imshow(img_before_rgb)
    # axes[i][0].set_title(f"[{i+1}] Before HSV")
    axes[i][0].axis("off")

    axes[i][1].imshow(img_after_rgb)
    # axes[i][1].set_title(f"[{i+1}] After HSV")
    axes[i][1].axis("off")

# plt.suptitle("색상 증강 전후 비교 (왼쪽: 전, 오른쪽: 후)", fontsize=16)
plt.tight_layout()
plt.show()
```
